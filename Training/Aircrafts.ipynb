```python
import csv
latitude={}
longitude={}
f = open("airports.dat")
for row in csv.reader(f):
    latitude[row[0]]= float(row[6])
    longitude[row[0]]= float(row[7])
```

```python
import geo_distance
f = open("routes.dat")
```

```python
routedistance = []
for row in csv.reader(f):
    if row[3] in latitude and row[5] in latitude:
        a = latitude[row[3]]
        b = longitude[row[3]]
        c = latitude[row[5]]
        d = longitude[row[5]]
        routedistance.append(geo_distance.distance(a, b, c, d))    
routedistance
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.hist(routedistance, 100, facecolor='r')
plt.xlabel("Distance (km)")
plt.ylabel("Number of flights")
```
